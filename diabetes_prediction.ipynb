```python
import numpy as np
import pandas as pd
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
```

```python
data_set=pd.read_csv("/content/diabetes.csv")
```

```python
print(data_set.head())
```

```python
data_set.describe()
```

```python
data_set["Outcome"].value_counts()
```

```python
data_set.groupby("Outcome").mean()
```

```python
X=data_set.drop(columns="Outcome",axis=1)
Y=data_set["Outcome"]
```

```python
print(X)
print(Y)
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=.1,stratify=Y,random_state=1)
```

```python
model=LogisticRegression()
```

```python
model.fit(X_train,Y_train)
```

```python
X_train_prediction=model.predict(X_train);
training_data_accuracy=accuracy_score(X_train_prediction,Y_train)
```

```python
print(training_data_accuracy)
```
